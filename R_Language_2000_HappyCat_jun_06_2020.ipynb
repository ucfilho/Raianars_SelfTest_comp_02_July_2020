```r
library('DEoptimR')

HappyCat = function(x)
  {
   Alpha=1. / 8
   Num=length(x)
   s=0
   for( i in 1:Num){s=s+x[i]*x[i]}
   fun=((s - length(x))**2)**Alpha +   
      (s / 2 + sum(x)) / length(x) + 0.5 
   return(fun )
    
  }
# global mimima at (-1,-1,...,-1) - f(x)=0  - d=[-2,2] (qlq dominio)
dim=30
RUNS=50
Y=0;X=0
for(i in 1:RUNS)
  {
  JDE_R=JDEoptim(rep(-2, dim), rep(2, dim), HappyCat,
         tol = 1e-20, NP=100, trace = FALSE, maxiter =2000)
  Y[i]=JDE_R$value
  X=JDE_R$par
  }
print(Y)
MEAN=mean(Y)
STD=sd(Y)
MAX=max(Y)
MIN=min(Y)
cat('HappyCat JDE DIM=',dim,'RUNS=',RUNS,'\n')
cat('MEAN=',MEAN,'\n')
cat('MAX',MAX,'\n')
cat('MIN=',MIN,'\n')
cat('STD',STD,'\n')
```
